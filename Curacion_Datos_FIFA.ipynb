```python
def clean_weak_foot(value):
    value = str(value).strip()

    base_value = int(value[0])

    if '+' in value:
        modifier = int(value.split('+')[1].strip())
        return min(base_value + modifier, 5)
    elif '-' in value:
        modifier = int(value.split('-')[1].strip())
        return max(base_value - modifier, 1)
    else:
        return base_value

df_24['Weak foot'] = df_24['Weak foot'].apply(clean_weak_foot)
df_24['Weak foot'].unique()
```

```python
posiciones = {
    # Arqueros
    'GK': 'Arquero',

    # Defensores
    'CB': 'Defensor', 'LB': 'Defensor', 'RB': 'Defensor',
    'LWB': 'Defensor', 'RWB': 'Defensor',

    # Mediocampistas
    'CDM': 'Mediocampista', 'CM': 'Mediocampista', 'CAM': 'Mediocampista',
    'LM': 'Mediocampista', 'RM': 'Mediocampista',

    # Delanteros
    'LW': 'Delantero', 'RW': 'Delantero', 'CF': 'Delantero', 'ST': 'Delantero'
}

def reductor_categorias(position_str):

    if pd.isna(position_str):
        return ['Unknown']

    positions = [pos.strip() for pos in position_str.split(',')]
    categories = list({posiciones.get(pos, 'Unknown') for pos in positions})

    return categories

df_24['Position'] = df_24['Position'].apply(reductor_categorias)
df_25['Position'] = df_25['Position'].apply(reductor_categorias)

print("\nDistribución de categorías en FIFA 24:")
print(df_24['Position'].explode().value_counts(normalize=True))

print("\nDistribución de categorías en FIFA 25:")
print(df_25['Position'].explode().value_counts(normalize=True))
```

```python
!pip install fuzzywuzzy
!pip install rapidfuzz

from rapidfuzz import fuzz, process
import re

def normalize_name(name):
    if pd.isna(name):
        return ""
    name = str(name).lower()
    return " ".join(sorted(re.sub(r'[^a-z\s]', '', name).split()))

def preprocess_names(names):
    return np.array([normalize_name(n) for n in names])

def find_common_players(df1, df2, name_col='Name', threshold=85):
    names1 = df1[name_col].dropna().unique()
    names2 = df2[name_col].dropna().unique()

    norm1 = preprocess_names(names1)
    norm2 = preprocess_names(names2)

    name_to_index = {name: idx for idx, name in enumerate(names2)}

    matches = []
    for idx, (orig_name, norm_name) in enumerate(zip(names1, norm1)):
        match = process.extractOne(
            norm_name,
            norm2,
            scorer=fuzz.token_set_ratio,
            score_cutoff=threshold
        )
        if match:
            matched_name = names2[match[2]]
            matches.append((orig_name, matched_name, match[1]))

    return pd.DataFrame(matches, columns=['FIFA24_Name', 'FIFA25_Name', 'Match_Score'])

common_players = find_common_players(df_24, df_25)

print(f"Jugadores en común: {len(common_players)}")
print(common_players.sort_values('Match_Score', ascending=False).head(10))

total_players = len(set(df_24['Name']) | set(df_25['Name']))
coverage = len(common_players) / total_players * 100
print(f"\nCobertura: {coverage:.2f}% de jugadores encontrados en ambos dataframes")

def find_matches(names_24, names_25, threshold=95):
    norm_24 = [normalize_name(n) for n in names_24]
    norm_25 = {normalize_name(n): i for i, n in enumerate(names_25)}

    matches = set()
    for name_24, norm_name in zip(names_24, norm_24):
        best_match = process.extractOne(
            norm_name,
            list(norm_25.keys()),
            scorer=fuzz.token_set_ratio,
            score_cutoff=threshold
        )
        if best_match:
            original_idx = norm_25[best_match[0]]
            matches.add(original_idx)

    return matches

names_24 = df_24['Name'].dropna().unique()
names_25 = df_25['Name'].dropna().unique()
matching_indices = find_matches(names_24, names_25)

df_fifa25_unique = df_25.drop(index=list(matching_indices)).reset_index(drop=True)

print(f"Jugadores originales en FIFA 25: {len(df_25)}")
print(f"Jugadores eliminados (presentes en FIFA 24): {len(matching_indices)}")
print(f"Jugadores únicos en FIFA 25: {len(df_fifa25_unique)}")
```
